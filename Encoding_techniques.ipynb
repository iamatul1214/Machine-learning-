## Here we have implemented different categorical encoding techniques on the titanic dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Titanic.csv')
df.head(10)
```

```python
PClassDict =   {   1:'First',
                    2:'Second',
                     3:'Third',              
                }
```

```python
## changing the pclass column into categorical
df['Pclass'] = df.Pclass.map(PClassDict)
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Age'].isna().sum()
```

```python
df_na = df.loc[df['Age'].isna()]
df_na
```

```python
df_na['Age'].fillna(df['Age'].mean(), inplace=True)
df_na
```

```python
df['Cabin'].mode()
```

```python
## Let's drop the missing values rows
print(df.shape)
df1 = df.dropna(axis=0, how='any')
print(df1.shape)
df1
```

```python
## We observe that dropping will lead to major data loss, hence fill the cabin NAN values with the mode values
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)
df.isna().sum()
```

### 2. Ordinal Encoding 

```python
## Here we will create a dictionary and then map it to the categorical columns
PClassDict =   {   'First':1,
                    'Second':2,
                     'Third':3,              
                }
df['Ordninal_Pclass'] = df.Pclass.map(PClassDict)
df
```

### 3. Count/Frequency Encoding

```python
Pclass_dict = df['Pclass'].value_counts()
df['Frequency_Encoded_Pclass'] = df.Pclass.map(Pclass_dict)
df
```

```python
Cabin_dict = df['Cabin'].value_counts()
df['Encoded_cabin'] = df.Cabin.map(Cabin_dict)
df
```

### 4. Target/Guided encoding

```python
df['Cabin_first_letter'] = df['Cabin'].astype(str).str[0]   ## create a new column and store the first letter of the cabin values
df.head(6)
```

```python
Encoded_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
Encoded_labels
```

```python
Encoded_labels_rank = {k:i for i,k in enumerate(Encoded_labels,0)}
Encoded_labels_rank
```

```python
df['Cabin_target_encoded'] = df.Cabin.map(Encoded_labels_rank)
df.head(5)
```

```python
Encoded_pClass = df.groupby(['Pclass'])['Survived'].mean().sort_values().index
Encoded_pClass
```

```python
Encoded_Pclass_dict = {k:i for i,k in enumerate(Encoded_pClass,0)}
Encoded_Pclass_dict
```

```python
df['Target_encoded_Pclass'] = df.Pclass.map(Encoded_Pclass_dict)
df.head(5)
```

### 5. Mean Encoding

```python
Encoded_Mean_Dict = df.groupby(['Cabin'])['Survived'].mean().to_dict()
Encoded_Mean_Dict
```

```python
df['Cabin_Mean_encoded'] = df.Cabin.map(Encoded_Mean_Dict)
df.head(5)
```

### 5. Probability encoding --> It is mostly for the binary classification

```python
Probability_survived = df.groupby(['Cabin_first_letter'])['Survived'].mean()
Probability_survived = pd.DataFrame(Probability_survived)
Probability_survived['Died'] = 1- Probability_survived
Probability_survived['Prob_ratio'] = Probability_survived['Survived']/Probability_survived['Died']
Probability_survived
```

```python
Encode_Prob_ratio_cabin = Probability_survived['Prob_ratio'].to_dict()
Encode_Prob_ratio_cabin
```

```python
df['Encoded_Prob_ratio_cabin'] = df.Cabin_first_letter.map(Encode_Prob_ratio_cabin)
df
```
